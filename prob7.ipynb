```python
def tower_of_hanoi(n, source, auxiliary, destination):
    if n == 1:
        print(f"Move disk 1 from {source} → {destination}")
        return
    
    # Move n-1 disks from source to auxiliary
    tower_of_hanoi(n-1, source, destination, auxiliary)

    # Move the nth (largest) disk to destination
    print(f"Move disk {n} from {source} → {destination}")

    # Move n-1 disks from auxiliary to destination
    tower_of_hanoi(n-1, auxiliary, source, destination)

# ----------- Run the Program -----------
n = int(input("Enter number of disks: "))
print("\nSteps to solve Tower of Hanoi:\n")
tower_of_hanoi(n, 'A', 'B', 'C')
```

```python
import time
import os

def clear():
    os.system('cls' if os.name == 'nt' else 'clear')

def print_rods(rods, n):
    clear()
    print()
    for level in range(n-1, -1, -1):
        for rod in ['A', 'B', 'C']:
            if level < len(rods[rod]):
                disk = rods[rod][level]
                print(f"  [{disk}]  ", end=" ")
            else:
                print("   |   ", end=" ")
        print()
    print(" ------  ------  ------")
    print("   A       B       C   ")
    print()

def move(n, source, auxiliary, destination, rods, total_disks):
    if n == 1:
        rods[destination].append(rods[source].pop())
        print_rods(rods, total_disks)
        time.sleep(0.7)
        return

    move(n-1, source, destination, auxiliary, rods, total_disks)
    rods[destination].append(rods[source].pop())
    print_rods(rods, total_disks)
    time.sleep(0.7)
    move(n-1, auxiliary, source, destination, rods, total_disks)


# ----------- RUN GAME -----------
n = int(input("Enter number of disks: "))

rods = {
    'A': list(range(n, 0, -1)),  # largest → smallest
    'B': [],
    'C': []
}

print_rods(rods, n)
move(n, 'A', 'B', 'C', rods, n)
print("\nCompleted Successfully!\n")
```
