```python
from collections import deque

def bfs_shortest_path(grid, start, goal):
    rows, cols = len(grid), len(grid[0])
    
    # Directions: Up, Down, Left, Right
    directions = [(0,1), (0,-1), (1,0), (-1,0)]
    
    # Queue for BFS â†’ stores (row, col, path)
    queue = deque([(start[0], start[1], [start])])
    
    visited = set()
    visited.add(start)

    while queue:
        r, c, path = queue.popleft()

        # If goal reached, return path
        if (r, c) == goal:
            total_cost = len(path) - 1
            return path, total_cost
        
        # Check all neighbors
        for dr, dc in directions:
            nr, nc = r + dr, c + dc
            
            # Check boundary + not a wall + not visited
            if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] == 0 and (nr, nc) not in visited:
                visited.add((nr, nc))
                queue.append((nr, nc, path + [(nr, nc)]))

    return None  # No path found

# ---------------- Example Maze ----------------
maze = [
    [0, 0, 1, 0],
    [1, 0, 1, 0],
    [0, 0, 0, 0],
    [0, 1, 1, 0]
]

start = (0, 0)   # Starting cell
goal = (3, 3)    # Ending cell

path, cost = bfs_shortest_path(maze, start, goal)

print("\nMaze Shortest Path:")
print("Path:", path)
print("Total Cost:", cost)
```
