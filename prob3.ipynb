```python
game_map = {
    "Camp": ["Forest", "River"],
    "Forest": ["Treasure Room"],
    "River": ["Treasure Room"],
    "Treasure Room": []
}

start_location = "Camp"
target_location = "Treasure Room"

# DFS Traversal Order
visited = set()

def dfs_traverse(node):
    print("Visiting:", node)
    visited.add(node)
    for next_node in game_map[node]:
        if next_node not in visited:
            dfs_traverse(next_node)

print("ORDER OF TRAVERSAL:\n")
dfs_traverse(start_location)


# DFS All Possible Paths
def dfs_paths(current, goal, path=[]):
    path = path + [current]
    
    if current == goal:
        return [path]
    
    paths = []
    for next_node in game_map[current]:
        if next_node not in path:
            new_paths = dfs_paths(next_node, goal, path)
            for p in new_paths:
                paths.append(p)
    return paths

print("\nALL POSSIBLE PATHS TO TREASURE:\n")
paths = dfs_paths(start_location, target_location)

for p in paths:
    print(" -> ".join(p))
```
