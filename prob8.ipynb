```python
from collections import deque

def water_jug_bfs(capA, capB, goal):
    visited = set()
    queue = deque()

    queue.append((0, 0))   # Start with both jugs empty
    visited.add((0, 0))

    while queue:
        A, B = queue.popleft()
        print(f"State: ({A}, {B})")

        # Check goal condition
        if A == goal or B == goal:
            print("\nGoal Reached!")
            print(f"Final State: ({A}, {B})")
            return
        
        # All possible next actions:
        next_states = [
            (capA, B),   # Fill Jug A
            (A, capB),   # Fill Jug B
            (0, B),      # Empty Jug A
            (A, 0),      # Empty Jug B
        ]

        # Pour A -> B
        transfer = min(A, capB - B)
        next_states.append((A - transfer, B + transfer))

        # Pour B -> A
        transfer = min(B, capA - A)
        next_states.append((A + transfer, B - transfer))

        # Add new states to queue
        for state in next_states:
            if state not in visited:
                visited.add(state)
                queue.append(state)

    print("\nNo solution found.")


# ---------------- RUN ----------------
capA = int(input("Enter capacity of Jug A: "))
capB = int(input("Enter capacity of Jug B: "))
goal = int(input("Enter the goal amount: "))

print("\n--- BFS Water Jug Solver ---\n")
water_jug_bfs(capA, capB, goal)
```
