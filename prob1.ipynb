```python
from queue import PriorityQueue

# Graph (Cities with distances in km)
graph = {
    'Delhi': {'Agra': 200, 'Mumbai': 900},
    'Agra': {'Delhi': 200, 'Gwalior': 120},
    'Gwalior': {'Agra': 120, 'Bhopal': 350},
    'Bhopal': {'Gwalior': 350, 'Indore': 180},
    'Indore': {'Bhopal': 180, 'Mumbai': 600},
    'Mumbai': {}
}

# Heuristic (Straight-line estimates towards Mumbai)
heuristic = {
    'Delhi': 1150,
    'Agra': 1000,
    'Gwalior': 850,
    'Bhopal': 600,
    'Indore': 400,
    'Mumbai': 0
}

def a_star_search(start, goal):
    pq = PriorityQueue()
    pq.put((heuristic[start], 0, start, [start]))   # (f, g, node, path)
    visited = set()

    while not pq.empty():
        f, cost, current, path = pq.get()

        if current in visited:
            continue
        visited.add(current)

        if current == goal:
            print("\nGoal Reached:", goal)
            print("Optimal Path:", " ‚Üí ".join(path))
            print("Total Travel Distance (Path Cost):", cost, "km")
            return

        for neighbor, edge_cost in graph[current].items():
            if neighbor not in visited:
                g = cost + edge_cost
                f = g + heuristic[neighbor]
                pq.put((f, g, neighbor, path + [neighbor]))

    print("\nNo path exists from", start, "to", goal)

# Run A* Search
a_star_search('Agra', 'Mumbai')
```
